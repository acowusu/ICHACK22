```python
post_codes = pd.read_csv('../input/post-code-to-coordinate/ukpostcodes.csv')
purchase_price = pd.read_csv('../input/purchase-prices/pp-2021.csv', usecols = [' postcode', ' price'])

length = len(post_codes.index)

purchase_price.sort_values(by = [' postcode'])
cnt = 1
sum = 0
post_code = ""
sales = {}
for i in range(1, len(purchase_price.index)):
    if removespace(str(purchase_price[' postcode'][i])) != removespace(str(purchase_price[' postcode'][i-1])): 
        sales[post_code] = sum // cnt
        sum = 0
        cnt = 0
    cnt += 1
    sum += purchase_price[' price'][i]
    post_code = removespace(str(purchase_price[' postcode'][i]))
        
coordinate_average_price = []
latitudes = []
longitudes = []
prices = []
for i in range(length):
    if removespace(post_codes['postcode'][i]) in sales:
        average_price = sales[removespace(post_codes['postcode'][i])]
        latitudes.append(post_codes['latitude'][i])
        longitudes.append(post_codes['longitude'][i])
        prices.append(average_price)
        current_coordinate = (post_codes['latitude'][i], post_codes['longitude'][i])
        coordinate_average_price.append((current_coordinate, average_price))

finaldata = {'latitude' : latitudes, 'longitude' : longitudes, 'price' : prices}

coordinate_price = df = pd.DataFrame(data=finaldata)
print(coordinate_price)

    
print(len(coordinate_average_price))
```
